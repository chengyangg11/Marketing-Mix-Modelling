```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('conjura_mmm_data.csv')
```

```python
df.columns
```

```python
# Example: find brand with longest available data
brand_counts = df.groupby('ORGANISATION_ID')['DATE_DAY'].count()
brand_to_use = brand_counts.idxmax()
brand_df = df[df['ORGANISATION_ID'] == brand_to_use]
```

```python
brand_df
```

```python
# Ensure DATE_DAY is datetime
brand_df['DATE_DAY'] = pd.to_datetime(brand_df['DATE_DAY'])

# Unique dates descending
unique_dates_desc = np.sort(brand_df['DATE_DAY'].unique())[::-1]

# Print all dates
for date in pd.to_datetime(unique_dates_desc):
    print(date.strftime('%Y-%m-%d'))
    
# Calculate approximate number of years
earliest_date = np.min(unique_dates_desc)
latest_date = np.max(unique_dates_desc)

# Convert timedelta64 to number of days
num_days = (latest_date - earliest_date) / np.timedelta64(1, 'D')
num_years = num_days / 365
print(f"Approximate number of years of data: {num_years:.2f} years")
```
